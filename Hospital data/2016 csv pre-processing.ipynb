```python
condition = df['APR Risk of Mortality'] == 'Minor'
df1 = df[condition]
```

```python
#filtering by potential respiratory conditions relating to pollution in the 'APR MDC Description' column, 
#resetting index

filtered = ['Diseases and Disorders of the Respiratory System']

condition = df1['APR MDC Description'].isin(filtered)
df2 = df1[condition]
df2 = df2.reset_index(drop = True)
```

```python
#dropping irrelevant columns
dropping_columns = [ 'Operating Certificate Number', 'Facility Id', 'Race', 'Ethnicity', 'Patient Disposition','CCS Procedure Code',
      'CCS Diagnosis Code', 'CCS Procedure Description', 'APR DRG Code','APR MDC Code', 'APR Medical Surgical Description',
                    'Emergency Department Indicator', 'Payment Typology 1',
       'Payment Typology 2', 'Payment Typology 3', 'Birth Weight','Abortion Edit Indicator',
      'Total Charges', 'Total Costs',]
df3 = df2.drop(columns = dropping_columns)
```
