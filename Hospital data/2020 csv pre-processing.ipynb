```python
condition = df['APR Risk of Mortality'] == 'Minor'
df1 = df[condition]
```

```python
#listing the categories in the MDC description

df1['APR MDC Description'].unique()
```

```python
#filtering by potential respiratory conditions relating to pollution in the 'APR MDC Description' column, 
#resetting index

filtered = ['DISEASES AND DISORDERS OF THE RESPIRATORY SYSTEM']

condition = df1['APR MDC Description'].isin(filtered)
df2 = df1[condition]
df2 = df2.reset_index(drop = True)
```

```python
df2.head()
```

```python
#dropping irrelevant columns
dropping_columns = [ 'Operating Certificate Number', 'Permanent Facility Id', 'Race', 'Ethnicity', 'Patient Disposition','CCSR Procedure Code',
      'CCSR Diagnosis Code', 'CCSR Procedure Description', 'APR DRG Code','APR MDC Code', 'APR Medical Surgical Description',
                    'Emergency Department Indicator', 'Payment Typology 1',
       'Payment Typology 2', 'Payment Typology 3', 'Birth Weight',
      'Total Charges', 'Total Costs',]
df3 = df2.drop(columns = dropping_columns)
```
